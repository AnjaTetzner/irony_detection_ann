```python
import numpy as np
import random as rd
import tensorflow as tf

from keras.layers import Dense, Embedding, Flatten, Input
from keras.models import Model
from keras.optimizers import Adagrad

from sklearn import preprocessing

#Import another jupyter notebook
import import_ipynb
from Productive_CrossValidation_inclTest import *
from Productive_Data_GetTrainTest import samEval
```

```python
#for reproducability of the results
seed = 7
np.random.seed(seed)
tf.set_random_seed(seed)
rd.seed(seed)
```

```python
path = 'data/'
outputpath ='outputs/03_withoutEmbeddings_samEval2018/'
```

```python
#Load necessary data
train_data, train_labels, test_data, test_labels, word_index, MAX_SEQUENCE_LENGTH = samEval()
```

```python
NUM_EPOCHS = 30
BATCH_SIZE = 64
optimizer = Adagrad(lr=0.001)
lossfunction = 'binary_crossentropy'
softmax = True
```

```python
#Input Layer
comment_input = Input(shape=(MAX_SEQUENCE_LENGTH,), dtype='float32')

#First Dense Layer
x = Dense(600, activation='relu')(comment_input)

#Output (Dense) Layer
preds = Dense(2, activation='softmax')(x)

# build the model
model = Model(inputs=[comment_input], outputs=preds)
model.summary()
```

```python
cross_val(model, NUM_EPOCHS, BATCH_SIZE, optimizer, lossfunction, train_data, train_labels, softmax, outputpath, 'withoutEmbedding')
```

## Normalize Data

```python
train_data_normalized = preprocessing.normalize(train_data)
```

```python
cross_val(model, NUM_EPOCHS, BATCH_SIZE, optimizer, lossfunction, train_data_normalized, train_labels, softmax, outputpath, 'withoutEmbedding')
```

## Data standadization

```python
train_data_standardized = preprocessing.scale(train_data)
```

```python
cross_val(model, NUM_EPOCHS, BATCH_SIZE, optimizer, lossfunction, train_data_standardized, train_labels, softmax, outputpath, 'withoutEmbedding_standardized')
```

# Test the best model with the heldout test data set

```python
test_data_standardized = preprocessing.scale(test_data)
train_evaluate_final_model(model, NUM_EPOCHS, BATCH_SIZE, optimizer, lossfunction, train_data_standardized, train_labels, test_data_standardized, test_labels, softmax, outputpath, 'final_withoutEmbedding')
```
